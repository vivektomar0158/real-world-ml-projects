## Fixing the skewed columns

```python
from sklearn.preprocessing import QuantileTransformer
from scipy.stats import skew

num_cols = df.select_dtypes(include='number').columns

skew_vals = df[num_cols].apply(skew)
skewed_cols = skew_vals[abs(skew_vals) > 1].index

qt = QuantileTransformer(output_distribution='normal', random_state=42)

for col in skewed_cols:
    df[[col]] = qt.fit_transform(df[[col]])
    print(f"Fixed skew in: {col}")
```

## Applying Standard Scaler 

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df.select_dtypes(include='number'))

df_scaled = pd.DataFrame(scaled_data, columns=df.select_dtypes(include='number').columns)
```

```python
print(df_scaled.mean().round(2))
print(df_scaled.std().round(2))
```

### Using elbow method to find Clusters

```python
from sklearn.cluster import KMeans

inertia = []
k_range = range(1,11)

for k in k_range: 
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,4))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

## Applying Kmeans

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(df_scaled)
```

```python
df['Cluster'] = kmeans.labels_ 
df['Cluster'].value_counts()
```

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_scaled)

pca_df = pd.DataFrame(data=pca_data, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = df['Cluster']
```

### Finding Segmentation

```python
#visualising 
for cluster in sorted(pca_df['Cluster'].unique()):
    cluster_data = pca_df[pca_df['Cluster'] == cluster]
    plt.scatter(cluster_data['PCA1'], cluster_data['PCA2'], label=f'Cluster {cluster}', s=30)

plt.title('Customer Segmentation (PCA Projection)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by cluster
cluster_profiles = df.groupby('Cluster').mean(numeric_only=True)
cluster_profiles
```

```python
# Group by cluster
numeric_cols = df.select_dtypes(include='number').columns.drop('Cluster')

cluster_profiles = df.groupby('Cluster')[numeric_cols].mean()

print(cluster_profiles)
```

## Each cluster Average Values

```python
top_features = ['BALANCE', 'PURCHASES', 'PAYMENTS', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES']

# Plot
cluster_profiles[top_features].T.plot(kind='bar', figsize=(10, 6))
plt.title("Average Feature Values by Cluster")
plt.xlabel("Features")
plt.ylabel("Average Value")
plt.legend(title="Cluster")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Conclusion

I worked on a credit card customer segmentation project using unsupervised learning, with the goal of identifying distinct customer groups based on their transaction behavior. The dataset was sourced from Kaggle and included features like balance, purchases, payments, and spending styles.

The curve clearly bent at k = 4, so I applied K-Means clustering with 4 clusters and added those labels back into the dataset.

For visualization, I used PCA (Principal Component Analysis) to reduce the feature space to 2 dimensions. The resulting 2D scatter plot showed well-separated clusters, confirming the effectiveness of the segmentation.
